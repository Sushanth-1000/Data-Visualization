```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('traffic_accidents_filtered.csv')

# Fill missing values in 'crash_month' with 0 before converting to int
df['crash_month'] = df['crash_month'].fillna(0).astype(int)

monthly = df.groupby('crash_month').size()

monthly.plot(kind='bar')
plt.title('Total Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()
```
