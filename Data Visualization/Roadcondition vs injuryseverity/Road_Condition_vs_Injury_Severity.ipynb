```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/traffic_accidents_filtered.csv')
pd.crosstab(df['roadway_surface_cond'], df['most_severe_injury']).plot(kind='bar', stacked=True)
plt.title('Road Condition vs Injury Severity')
plt.ylabel('Number of Accidents')
plt.show()
```
