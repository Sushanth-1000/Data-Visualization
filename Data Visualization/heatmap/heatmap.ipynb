```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('traffic_accidents_filtered.csv')

```

```python
df['crash_month'] = df['crash_month'].fillna(0).astype(int)
df['crash_day_of_week'] = df['crash_day_of_week'].fillna(0).astype(int)

# Create a pivot table for the heatmap
heatmap_data = df.pivot_table(index='crash_day_of_week', columns='crash_month', values='Date', aggfunc='count')

# Plot the heatmap
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Heatmap of Accidents by Day of Week and Month')
plt.xlabel('Month')
plt.ylabel('Day of Week')
plt.show()
```
