```python
import numpy as np
import pandas as pd
from dateutil import parser
```

```python
df=pd.read_csv('/content/traffic_accidents.csv')
df.head()
```

```python
TA=pd.DataFrame(df)
```

```python
TA.shape
```

```python
TA.isnull()
```

```python
TA.isnull().sum()
```

```python
TA.nunique()
```

```python
TA['Datetime'] = TA['crash_date'].apply(lambda x: parser.parse(x, fuzzy=True) if pd.notnull(x) else pd.NaT)
TA['Date']=TA['Datetime'].dt.date
TA['Time']=TA['Datetime'].dt.time
```

```python
TA.drop('Datetime',axis=1,inplace=True)
```

```python
TA.head()
```

```python
TA.shape
```

```python
TA2=TA.drop('crash_date',axis=1)
```

```python
TA2.head()
```

```python
cols = ['Date', 'Time'] + [col for col in TA2.columns if col not in ['Date', 'Time']]
TA2 = TA2[cols]
```

```python
TA2.to_csv('traffic_accidents_filtered.csv', index=False)
```
